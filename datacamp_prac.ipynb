Task:

Create a Pandas DataFrame with:

Day: [1, 2, 3, 4, 5]

Sales: [200, 450, 300, 500, 650]

Using Matplotlib:

Plot a line plot of Day vs Sales

Plot a scatter plot of Day vs Sales on the same figure

Add:

Title: "Sales Trend"

X-axis label: "Day"

Y-axis label: "Sales"

```python
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'day':[1,2,3,4,5],
    'sales':[200, 450, 300, 500, 650]
})
print(df)

df.plot(kind='scatter',x='day',y='sales')
plt.xlabel("day")
plt.ylabel("Sales")
plt.title("sales trends")
plt.show()
```
