```python
fliter_data=df[df['release_date']>='2015']
fliter_data
```

```python
short_movies=fliter_data[fliter_data['runtime']<=90]
short_movies
```

```python
vote_avg=short_movies[short_movies['vote_average']>7]
vote_avg
```

```python
# top_genre_recent_short_popular=vote_avg.iloc[0]
# top_genre_recent_short_popular
import ast
def q(r):
  if pd.isna(r):
    return []
  g_l=ast.literal_eval(r)
  return [g['name'] for g in r]
  vote_avg['g_n']=vote_avg['genres'].apply(q)
  all_gener=vote_avg['g_n'].explode()
  t=all_gener.value_counts.idxmax()
  print(t)
```
