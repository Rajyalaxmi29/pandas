```python
import pandas as pd
import numpy as np

data = {
    "id": [1, 2, 3, 4, 5, 6],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "department": ["HR", "IT", "IT", "Finance", "HR", None],
    "score": [85, 92, 78, 88, None, 95],
    "salary": [50000, 70000, 65000, None, 52000, 80000]
}

df = pd.DataFrame(data)
df
```
