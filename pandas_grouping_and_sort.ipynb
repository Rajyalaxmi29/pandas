üü¢ QUESTION 2 (Easy ‚Üí Medium)

Find the average points given to wines from each country, and display the result sorted from highest average points to lowest.

üü¢ QUESTION 6 (HARD ‚Äì MultiIndex + reset_index + sorting)

Now we‚Äôll combine multiple concepts you just learned.

For each country and province combination, count how many wine reviews there are.
Then convert the result into a normal DataFrame and sort it by the count in descending order.

Expected steps (no code given):

Group by country AND province

Count rows

Reset index

Sort by count (descending)

```python
g=(df.groupby(['country','province']).points.count()
.reset_index(name='review_count').sort_values(by='review_count',ascending=False))
g
```

‚ùì Question 7 (FINAL)

Find the top 5 (country, province) combinations that have the highest average wine points.
Return a normal DataFrame sorted by average points (descending).

Expected thinking:

Group by country and province

Calculate mean points

Reset index

Sort by mean points

Take top 5

```python
r=(df.groupby(['country','province']).points.mean().reset_index(name='h_a')
.sort_values(by='h_a',ascending=False).head(5))
r
```

```python
import pandas as pd
df=pd.DataFrame({ 'employee_id': [101,102,103,104,105,106,107,108,109,110],
    'name': ['Alice','Bob','Charlie','David','Eva','Frank','Grace','Helen','Ian','Jane'],
    'department': ['IT','HR','IT','Finance','HR','IT','Finance','HR','IT','Finance'],
    'experience_years': [5,3,8,10,2,6,4,7,1,9],
    'monthly_salary': [70000,45000,90000,120000,None,75000,65000,72000,40000,110000],
    'performance_score': [88,82,91,94,79,85,88,90,76,92],
    'location': ['Bangalore','Hyderabad','Bangalore','Mumbai','Hyderabad',
                 'Bangalore','Mumbai','Hyderabad','Bangalore','Mumbai']})
df
```

```python
df.loc[df.monthly_salary.notnull()]
```

```python
def q(r):
  if r.performance_score>=90:
     return "excellent"
  elif r.performance_score>=80:
    return "good"
  else:
    return "average"
df['Performance_Level']=df.apply(q,axis='columns')

df['value_index']=df.performance_score / df.monthly_salary
df
```

```python
s=df.groupby('department')[['performance_score','monthly_salary']].mean()
s
```

```python
s=df.groupby('department').performance_score.mean()
r=s.sort_values(ascending=False)
r
```

```python
r=df.groupby('name').value_index.max()
res=r.sort_values(ascending=False).head(3)
res

```
