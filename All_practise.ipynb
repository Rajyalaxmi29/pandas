```python
#random walk
np.random.seed(123)
r_w=[0]
for i in range(100):
  step=r_w[-1]
  d=np.random.randint(1,7)
  if d<=2:
    step=step-1
  elif d>=5:
    step=step+1
  else:
     step=step+random.randint(1,7)
  r_w.append(step)
print(r_w)
```
