Conditional Selection – AND (&)

Select all wines where country is "Italy" AND points are greater than or equal to 90.

Select all wines where price is greater than or equal to 30 AND points are greater than or equal to 95.

Select all wines where country is "France" AND price is not null.

```python
df.loc[(df.country=='Italy')&(df.points>=90)]
```

```python
df.loc[(df.price>=30)&(df.points>=95)]
```

```python
df.loc[(df.country=='France')&(df.price.notnull())]
```
